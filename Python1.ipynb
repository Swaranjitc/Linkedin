# Visual Analysis
Univariate, Bivariate and Multivariate Analysis

```python
#Which department has the most Chrun?
df.dtypes
```

```python
df.groupby('Departments')['left'].sum().plot(kind='bar')
```

```python
plt.style.use('ggplot')
plt.figure(figsize= (8,5))
df.groupby('Departments')['left'].sum().plot(kind='barh')
```

```python
df.groupby('Departments')['left'].sum().plot()
```

```python
plt.figure(figsize= (6,5))
plt.scatter(df['satisfaction_level'],df['last_evaluation'],c='Magenta', marker = '*')
plt.title("Bivariate Analysis: Satisfaction Vs Last Evaluation")
plt.savefig("Correlation.png");
```

```python
plt.figure(figsize= (6,5))
sns.distplot(df['satisfaction_level'])
```

```python
plt.figure(figsize= (12,5))
sns.boxplot(x=df['Departments'],y=df['satisfaction_level'])
```

```python
sns.heatmap(df.corr(), annot= True, linewidths=0.5)
```

```python
plt.figure(figsize= (12,5))
sns.violinplot(x=df['salary'],y=df['satisfaction_level'])
```
